```python
#20L-1186  20L-2156

import copy

def pri(puzzle):
    for i in range(3):
        for j in range(3):
            print(puzzle[i][j], "\n")
        print()
    print()

def move_check(x,y):#checks wether move is valid or not
  return x>-1 and y>-1 and x<3 and y<3 

def next_move(puzzle, i, j, x, y):
    puzzle[i][j], puzzle[x][y] = puzzle[x][y], puzzle[i][j]

def goal(puzzle):
    goal1 = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    goal2 = [[0, 1, 2], [3, 4, 5],[ 6, 7, 8]]
    return puzzle == goal1 or puzzle == goal2

def ids(visited,stack,puzzle,max_depth=10):
  if max_depth>0:
    i=0 
    level=0

    while(len(stack)):
     #print("Level : ",level)
     while(i<=level):
       puzzle,path=stack.pop()
       i=i+1
       if goal(puzzle):
        # solution found
        print("Solution Found at level ",level);
         # printing path
        for i in path:
          for j in range(3):
             print(i[j])
          print("\n")  
        print("\n")   
        return
       elif puzzle not in visited:
        l=copy.deepcopy(puzzle)
        visited.append(l) 
        i=i+1

        #finding blank position       
        x, y = 0, 0
        for i in range(3):
         for j in range(3):
            if puzzle[i][j] == 0:
                x, y = i, j
                break
        
        #move up
        if move_check(x-1,y):
         m=copy.deepcopy(puzzle)
         next_move(m, x, y, x - 1, y)

         if m not in visited:
           l=copy.deepcopy(m)
           currentMoveToPath=path[:]+[m]
           #append node to stack
           stack.append((l,currentMoveToPath))
           
        #move down 
        if move_check( x + 1, y):
         m=copy.deepcopy(puzzle)
         next_move(m, x, y, x + 1, y)
         if m not in visited:
           l=copy.deepcopy(m)
           currentMoveToPath=path[:]+[m]
           #append node to stack
           stack.append((l,currentMoveToPath))

        #move left          
        if move_check( x, y - 1):
         m=copy.deepcopy(puzzle)
         next_move(m, x, y, x, y - 1)
         if m not in visited:
           l=copy.deepcopy(m)
           currentMoveToPath=path[:]+[m]
           #append node to stack
           stack.append((l,currentMoveToPath))

        #move right          
        if move_check(x, y + 1):
         m=copy.deepcopy(puzzle)
         next_move(m, x, y, x, y + 1)
         if m not in visited:
          l=copy.deepcopy(m)
          currentMoveToPath=path[:]+[m]
          #append node to stack
          stack.append((l,currentMoveToPath))
       else:
          None

     level=level+1     
    print("Solution not found")

#main function
puzzle =  [[1,2,3],
          [4,5,6], 
          [0,7,8]]
visited=[]
b=copy.deepcopy(puzzle)
stack=[(b,[b])] # empty array for path
ids(visited,stack,puzzle)



```
