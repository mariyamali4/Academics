```python
# Q1: Part 1: Best First Search

import copy
import heapq

def move_check(x,y):#checks wether move is valid or not
  return x>-1 and y>-1 and x<3 and y<3 

def next_move(puzzle, i, j, x, y):
    new_puzzle = copy.deepcopy(puzzle)
    new_puzzle[i][j], new_puzzle[x][y] = new_puzzle[x][y], new_puzzle[i][j]
    return new_puzzle


def createMoves(puzzle):
    # Finding blank position
    x, y = 0, 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == 0:
                x, y = i, j
                break

    # Generate child states by swapping the blank tile with adjacent tiles
    child = []
    # Move up
    if x > 0:
        p = next_move(puzzle, x, y, x - 1, y)
        child.append(p)
    # Move down
    if x < 2:
        p = next_move(puzzle, x, y, x + 1, y)
        child.append(p)
    # Move left
    if y > 0:
        p = next_move(puzzle, x, y, x, y - 1)
        child.append(p)
    # Move right
    if y < 2:
        p = next_move(puzzle, x, y, x, y + 1)
        child.append(p)
    return child


def goal(puzzle):
    goal1 =  [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    return puzzle == goal1

def printPath(path):
 print("Solution Found")
 for i in path:
  for j in range(3):
    print(i[j]) 
  print("\n")  

def Heuristic(puzzle):
    h = 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] != 0:
                # compute the Manhattan distance of the tile
                goal_i, goal_j = (puzzle[i][j]-1)//3, (puzzle[i][j]-1)%3
                h += abs(i - goal_i) + abs(j - goal_j)
    return h

def BestFS(puzzle):

 b=copy.deepcopy(puzzle)
 visited=set()
 h=Heuristic(b)
 Heap=[(h,b,[b])] # [b] list for path
 solutionFound=0

 while Heap:

     h, puzzle, path = heapq.heappop(Heap)
     if goal(puzzle):
       printPath(path)
       solutionFound=1
       break
     visited.add(str(puzzle)) 
     for child in createMoves(puzzle):
         if str(child) not in visited:
           currentMoveToPath = path[:] + [child]
           #append node to Heap
           l = copy.deepcopy(child)
           h=Heuristic(l)
           heapq.heappush(Heap,((h,l, currentMoveToPath)))

 if solutionFound==0:
   print('Solution not found')


#-----------------------------------Main-----------------------------------------------


puzzle =  [[7,2,4],
           [5,0,6],
           [8,3,1]]
BestFS(puzzle)
```

```python
# Q1: Part 2: A* Search


import copy
import heapq

def move_check(x,y):#checks wether move is valid or not
  return x>-1 and y>-1 and x<3 and y<3 

def next_move(puzzle, i, j, x, y):
    new_puzzle = copy.deepcopy(puzzle)
    new_puzzle[i][j], new_puzzle[x][y] = new_puzzle[x][y], new_puzzle[i][j]
    return new_puzzle


def createMoves(puzzle):
    # Finding blank position
    x, y = 0, 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == 0:
                x, y = i, j
                break

    # Generate child states by swapping the blank tile with adjacent tiles
    child = []
    # Move up
    if x > 0:
        p = next_move(puzzle, x, y, x - 1, y)
        child.append(p)
    # Move down
    if x < 2:
        p = next_move(puzzle, x, y, x + 1, y)
        child.append(p)
    # Move left
    if y > 0:
        p = next_move(puzzle, x, y, x, y - 1)
        child.append(p)
    # Move right
    if y < 2:
        p = next_move(puzzle, x, y, x, y + 1)
        child.append(p)
    return child


def goal(puzzle):
    goal1 =  [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    return puzzle == goal1

def printPath(path):
 print("Solution Found");
 c = 0
 for i in path:
  print('Level ', c)
  c = c+1
  for j in range(3):
    print(i[j]) 
  print("\n")  

def Heuristic(puzzle):
    h = 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] != 0:
                # compute the Manhattan distance of the tile
                goal_i, goal_j = (puzzle[i][j]-1)//3, (puzzle[i][j]-1)%3
                h += abs(i - goal_i) + abs(j - goal_j)
    return h

def A_star(puzzle):

 b=copy.deepcopy(puzzle)
 visited=set()
 h=Heuristic(b)
 g=0
 Heap=[(h+g,b,[b])] # [b] list for path
 solutionFound=0

 while Heap:

     h, puzzle, path = heapq.heappop(Heap)
     if goal(puzzle):
       printPath(path)
       solutionFound=1
       break
     visited.add(str(puzzle)) 
     g=g+1 
     for child in createMoves(puzzle):
         if str(child) not in visited:
           currentMoveToPath = path[:] + [child]
           #append node to Heap
           l = copy.deepcopy(child)
           h=Heuristic(l)
           heapq.heappush(Heap,((h+g,l, currentMoveToPath)))

 if solutionFound==0:
   print('Solution not found')

#main function
puzzle =  [[7, 2, 4],
           [5, 0, 6],
           [8, 3, 1]]
A_star(puzzle)
```

```python
# Q2: MinMax



!pip install colorama
import copy
import colorama
from colorama import Fore, Back
from collections import deque as queue

#--------------------------------------------------------------------------------------
R = 'R'
B = 'B'
draw = 'D'

#--------------------------------------------------------------------------------------


def printboard(board):
  length = len(board) -1
  for i in range(len(board)):
    for j in range(len(board)):
      if (i == 0) and (j!=0 and j!=length):
        print(" ", Fore.RED + board[i][j], end = " ")

      elif (i == 0) and (j==length):
        print(" ", Fore.RED + board[i][j]+ '\n')

      elif (i == length) and (j!=0 and j!=length):
        print(" ", Fore.RED + board[i][j], end = " ")

      elif (j == 0) and (i!= 0 and i!=length):
        print(" ", Fore.BLUE + board[i][j], end = " ")

      elif (j== length) and (i!= 0 and i!=length):
        print(" ", Fore.BLUE + board[i][j]+ '\n')
        
      else:
        if (board[i][j] == '+'):
          print(" ", Fore.RED + board[i][j], end = " ")

        elif (board[i][j] == '-'):
          print(" ", Fore.BLUE + board[i][j], end = " ")

        else:
          print(" ", Fore.BLACK + board[i][j], end = " ")



#--------------------------------------------------------------------------------------

  

#--------------------------------------Utility Function-------------------------------------
def player(board):
    """
    Returns the player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j=='.':
                count += 1
    if count % 2 != 0:
      return "R"
    return "B"




def result(board, action):
    """
    Returns resulting board after any move
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    if curr_player==R:
     result_board[i][j] = "+"
    else:
     result_board[i][j] = "-"  
    return result_board



def terminal_node(board):
  for i in range(len(board)):
    for j in range(len(board)):
      if (board[i][j] == '.'):
        return False
  return True   


def evaluate_score(board):
  if getWinner(board) == R:
    return 1
  elif getWinner(board) == B:
    return -1
  else:
    return 0



#--------------------------------------------------------------------------------------

rowmoves = [-1, 1, 0, 0]
colmoves = [0, 0, -1, 1]

def validMove(arr, visited, x, y, player):
	if (player == 'R'):
		if x>=0 and x<len(arr) and y>=0 and y<len(arr[0]):
			if (arr[x][y] == '+') or (arr[x][y] == player):
				if visited[x][y] == False:
					return True

	elif (player == 'B'):
		if x>=0 and x<len(arr) and y>=0 and y<len(arr[0]):
			if (arr[x][y] == '-') or (arr[x][y] == player):
				if visited[x][y] == False:
					return True
	return False


def BFS(board, visited, row, col, player):

	q = queue()
	q.append(( row, col ))
	visited[row][col] = True

	while (len(q) > 0):
		cell = q.popleft()
		x = cell[0]
		y = cell[1]
		
		for i in range(4):
			nx = x + rowmoves[i]
			ny = y + colmoves[i]
			if (validMove(board, visited, nx, ny, player)):
				q.append((nx, ny))
				visited[nx][ny] = True

				if (board[nx][ny] == player and (nx == 6 or ny == 6)):
					return True

	return False

	
def redWinner(board):
	rows, cols = 7, 7
	visited = [[ False for i in range(cols)] for j in range(rows)]

	pathExists = False
	for col in range (1, 6):
		pathExists = BFS(board, visited, 0, col, 'R')

		if pathExists == True:
			break

	return pathExists



def blueWinner(board):
	rows, cols = 7, 7
	visited = [[ False for i in range(cols)] for j in range(rows)]

	pathExists = False
	for row in range (1, 6):
		pathExists = BFS(board, visited, row, 0, 'B')

		if pathExists == True:
			break

	return pathExists



def getWinner(board):
  if redWinner(board):
    print('\n\n\n Red wins!')
    return R

  elif blueWinner(board):
    print('\n\n\n Blue wins!')
    return B

  else:
    print('\n\n\n Draw!')
    return draw


#--------------------------------------------------------------------------------------



#--------------------------------------------------------------------------------------
def get_Red_Moves(board):
    r = []
    for i in range (len(board)):
        for j in range (len(board[0])):
            if board[i][j] == "R" or board[i][j]=="+":
                m = [(i+1, j), (i, j+1), (i-1, j), (i, j-1)] 

                for x, y in m:
                    if (0 < x < len(board)) and (0 < y < len(board[0]) ) and (board[x][y] == '.'):
                        r.append((x, y)) 
    moves = [*set(r)]
    moves.sort()
    return moves  


def get_Blue_Moves(board):
    b = []
    for i in range(len(board)):
        for j in range(len(board[0])):
            if board[i][j] == "B" or board[i][j]=="-":
                m = [(i+1, j), (i, j+1), (i-1, j), (i, j-1)] 
                for x, y in m:
                    if (0 < x < len(board)) and (0 < y < len(board[0])) and (board[x][y] == '.'):
                        b.append(( x, y)) 
    moves = [*set(b)]
    moves.sort()
    return moves   

#--------------------------------------------------------------------------------------



#-------------------------------------------------------------------------------------

def alpha_beta_pruining(board):
    
    alpha = float('-inf')
    beta = float('inf')
    curr_player = player(board)

    if curr_player == R:
        value, action = max_value(board, alpha, beta)
    else:
        value, action = min_value(board, alpha, beta)
    print("action in alpha_beta_pruining ",action)
    return action


#-------------------------------------------------------------------------------------

def min_value(board, alpha, beta):

    if terminal_node(board):
      score = evaluate_score(board)
      return score, None

    min_value = float('inf')
    min_indexes = None

    for action in get_Red_Moves(board):
        action_value, _ = max_value(result(board, action), alpha, beta)
        if int(action_value) < min_value:
            min_value = action_value
            min_indexes = action

        if min_value <= alpha:
            return min_indexes, min_indexes

        beta = min(beta, min_value)
    print("min_value",min_value)
    print("min_indexes",min_indexes)
    return min_value, min_indexes

#-------------------------------------------------------------------------------------

def max_value(board, alpha, beta):

    if terminal_node(board):
      print('terminal_Node',board)
      score = evaluate_score(board)
      return score, None

    max_value = float('-inf')
    max_indexes = None

    for action in get_Blue_Moves(board):
        action_value, _ = min_value(result(board, action), alpha, beta)
        print("action_value",action_value)
        print('max_value',max_value)
        if int(action_value) > max_value:
            max_value = action_value
            max_indexes = action

        if max_value >= beta:
            return max_value, max_indexes

        alpha = max(alpha, max_value)

    return max_value, max_indexes

#-------------------------------------------------------------------------------------  



def play_game(board):
    printboard(board)
    while not terminal_node(board):
        print("Enter the row and column number")
        row, col = map(int, input().split())
        board = result(board, (row, col))
        if not terminal_node(board):
            action = alpha_beta_pruining(board)
            board = result(board, action)
            printboard(board)
    print("Winner is ", getWinner(board))





#--------------------------------------------------------------------------------------
 

small_board = [[' ', 'R', 'R', 'R', 'R', 'R', ' '],
               ['B', '.', '.', '.', '.', '.', 'B'],
               ['B', '.', '.', '.', '.', '.', 'B'],
               ['B', '.', '.', '.', '.', '.', 'B'],
               ['B', '.', '.', '.', '.', '.', 'B'],
               ['B', '.', '.', '.', '.', '.', 'B'],
               [' ', 'R', 'R', 'R', 'R', 'R', ' ']
                ]


medium_board = [[' ', 'R', 'R', 'R', 'R', 'R', 'R', 'R', 'R', ' '],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],              
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                ['B', '.', '.', '.', '.', '.', '.', '.', '.', 'B'],
                [' ', 'R', 'R', 'R', 'R', 'R','R', 'R', 'R', ' ']
                ]


board = [[' ', 'R', 'R', 'R', 'R', 'R', ' '],
         ['B', '+', '.', '+', '.', '+', 'B'],
         ['B', '.', '+', '+', '.', '+', 'B'],
         ['B', '-', '-', '-', '+', '+', 'B'],
         ['B', '.', '.', '-', '-', '+', 'B'],
         ['B', '-', '-', '.', '-', '-', 'B'],
         [' ', 'R', 'R', 'R', 'R', 'R', ' ']
         ]

```
